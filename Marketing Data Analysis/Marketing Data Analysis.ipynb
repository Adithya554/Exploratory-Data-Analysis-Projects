```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1. Income
df['Yearly Income'].hist(bins=10)
plt.title("Income distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
#2.Age
df['Age'].hist(bins=10)
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
Graduate=Basic=none=Masters=PhD=0
for ed in df['Education']:
    if ed=="Graduation":
        Graduate=Graduate+1
    elif ed=="Masters":
        Masters=Masters+1
    elif ed=="PhD":
        PhD=PhD+1
    elif ed=="Basic":
        Basic=Basic+1
    else:
        none=none+1
Education=[none,Basic,Graduate,Masters,PhD]
levels=["None","Basic","Graduate","Masters","PhD"]
plt.bar(levels,height=Education)
plt.title("Education level distribution")
plt.xlabel("Level of Education")
plt.ylabel("Frequency");
```

```python
df1=df[['AmtRegularProds','AmtWines','AmtFishProducts','AmtFruits','AmtSweetProducts','AmtTotal']].copy()
dataplot = sns.heatmap(df1.corr(), cmap="YlGnBu", annot=True) 
```

```python
plt.scatter(df['AmtWines'], df['AmtRegularProds'])
z=np.polyfit(df['AmtWines'],df['AmtRegularProds'],1)
p=np.poly1d(z)
plt.plot(df['AmtWines'],p(df['AmtWines']))
plt.title("Relationship between amount spent on wines and amount spent on regular products")
plt.xlabel("Amount spent on wines")
plt.ylabel("Amount spent on regular products");
```

```python
plt.scatter(df['AmtWines'], df['AmtTotal'])
z=np.polyfit(df['AmtWines'],df['AmtTotal'],1)
p=np.poly1d(z)
plt.plot(df['AmtWines'],p(df['AmtWines']))
plt.title("Relationship between amount spent on wines and total amount spent.");
```

```python
meat=df['AmtMeatProducts'].sum()
fish=df['AmtFishProducts'].sum()
fruits=df['AmtFruits'].sum()
sweet=df['AmtSweetProducts'].sum()
regular=df['AmtRegularProds'].sum()
wine=df['AmtWines'].sum()
l=[meat,fish,fruits,sweet,regular,wine]
plt.pie(l,labels=["Meat","Fish","Fruits","Sweets","Regular","Wine"],autopct='%.2f',radius=2);
```
